```python
!pip install biopython
print()
from Bio import Align
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
wt_p53= Seq("CGCCCATCTCCTTCACC")
mut_p53= Seq("CGCCAATCTCCTTCACC")
## as we are trying to detect a point mutation, I deliberately changed only one base
wt_gc= gc_fraction(wt_p53) * 100
mut_gc= gc_fraction(mut_p53) * 100
## I multiplied 100 to get a whole number which is optional
Aligner= Align.PairwiseAligner()
Aligner.mode= 'global'
Alignment= Aligner.align(wt_p53, mut_p53)
print (Alignment[0])
print("Analysis Complete: Single Point Mutation (SNP) detected.")
```
